## Linear SVM


A CCD records magnitudes (which in astronomy is negatively related to brightness) in five different colors by taking images through five color filters. A filter is a kind of screen that blocks out all light except for light with a specific color. The SDSS telescope's filters are green (g), red (r), and three colors that correspond to light not visible to the human eye: ultraviolet (u), and two infrared wavelengths (i and z). The five magnitudes (through the five filters) of a star are thus symbolized by u, g, r, i, and z.

However, color in astronomy is symbolized by subtracting the magnitudes: u-g, g-r, r-i, and i-z (if we start from 5 filters). Remember that all these quantities involve magnitudes, so they decrease with increasing light output. Therefore the higher the value of g-r color the redder a star is.

On the thermal radiation curve example above (diagram above), the g intensity is greater than the r intensity therefore the g-r color will be negative (remember that magnitudes and intensity/brightness are negatively related) while the u-g color will be positive.

```python
XX = X[:,:2]
sv = LinearSVC()
sv.fit(X,y)
sv.fit(XX,y)
```

```python
sv.score(XX,y)
```

```python
sv.score(XX,y)
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(1,1, figsize = (10,10))
ax.scatter(XX[:,0], XX[:,1], c = y);
```

```python
x_L = np.linspace(.7, 1.34)
y_L = 1.2 -(sv.intercept_[0] / sv.coef_[0][1]) - (sv.coef_[0][0]/ sv.coef_[0][1])* x_L
```

```python
fig, ax = plt.subplots(1,1, figsize = (10,10))
ax.scatter(XX[:,0], XX[:,1], c = y)
ax.plot(x_L, y_L);
```
